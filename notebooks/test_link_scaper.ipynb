```python
# Input filing URL
import sys 
sys.path.append("../")
from helpers.table_scraper_version2 import *

filing_url = "https://www.sec.gov/ix?doc=/Archives/edgar/data/0001543151/000154315125000008/uber-20241231.htm"

# Initialize scraper
scraper = FinancialStatementScraper(filing_url)

# Extract all three statements at once
results = scraper.extract_all_statements(display_output=True)

# Summary
print("\n" + "="*80)
print("âœ… EXTRACTION SUMMARY")
print("="*80)

for stmt_type, result in results.items():
    status = result.get('status', 'unknown')
    name = result.get('statement_name', stmt_type)
    rows = result.get('rows_extracted', 0)
    output_file = result.get('output_file', 'N/A')

    print(f"\n{name}:")
    print(f"  Status: {status}")
    print(f"  Rows: {rows}")
    print(f"  File: {output_file}")

print("\n" + "="*80)
```
